```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. On nettoie les NaN dans 'brand' (crucial)
df_clean = df.dropna(subset=['brand'])

# 2. On identifie les Top 25 marques (pour que ce soit lisible)
top_brands = df_clean['brand'].value_counts().head(25).index

# 3. On filtre le dataset pour ne garder que ces marques
df_top = df_clean[df_clean['brand'].isin(top_brands)]

# 4. On crée une table "Ticket x Marque" (binaire : acheté ou non)
# On aggrege par ticket pour voir les co-occurrences
ticket_brand_matrix = pd.crosstab(df_top['anonymized_Ticket_ID'], df_top['brand']).astype(int)

# On binarise : si acheté > 0 -> 1
ticket_brand_matrix = (ticket_brand_matrix > 0).astype(int)

# 5. On calcule la matrice de corrélation (co-occurrence)
# C'est la méthode la plus rapide avec Pandas
brand_corr = ticket_brand_matrix.corr()

# 6. On trace la Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(brand_corr, 
            cmap='YlGnBu',  # Un dégradé clair vers foncé
            xticklabels=True, 
            yticklabels=True,
            vmax=0.3,       # On "cap" pour mieux voir les petites corrélations
            vmin=0)         # Évite le négatif pour la clarté

plt.title('Matrice de Co-occurrence : Probabilité d\'achat croisé entre Top Marques')
plt.tight_layout()

# Sauvegarde l'image pour ton PDF
plt.savefig('heatmap_marques_crossell.png')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. On nettoie les valeurs manquantes sur les colonnes clés
df_clean = df.dropna(subset=['brand', 'RFM_Segment_ID'])

# 2. On isole les 20 Top Marques pour que le visuel reste propre
top_brands = df_clean['brand'].value_counts().head(20).index
df_top = df_clean[df_clean['brand'].isin(top_brands)]

# 3. On crée le tableau croisé : Segment RFM (lignes) x Marques (colonnes)
# normalize='index' est la clé ici : on regarde la "part de marché" de la marque 
# AU SEIN de chaque segment (chaque ligne fait 100%).
rfm_brand_matrix = pd.crosstab(df_top['RFM_Segment_ID'], df_top['brand'], normalize='index') * 100

# 4. On trace la Heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(rfm_brand_matrix, 
            cmap='Oranges',   # Un beau dégradé (tu peux tester 'Reds' ou 'Purples')
            annot=False,      # On désactive les chiffres pour garder le graphique épuré
            linewidths=.5)

# 5. Esthétique du graphique
plt.title("Évolution des affinités de Marques selon le niveau de fidélité (Segment RFM)", fontsize=14, fontweight='bold')
plt.xlabel("Top 20 Marques", fontsize=12)
plt.ylabel("Segment RFM", fontsize=12)
plt.xticks(rotation=45, ha='right') # On incline les noms des marques pour la lisibilité
plt.tight_layout()

# Sauvegarde pour ton PDF
plt.savefig('heatmap_rfm_marques.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. PRÉPARATION DES DONNÉES COMMUNES ---
df_clean = df.dropna(subset=['brand', 'RFM_Segment_ID']).copy()
df_filtered = df_clean[
    (df_clean['RFM_Segment_ID'] != 99999) & 
    (~df_clean['brand'].isin(['SEPHORA', 'GIFT CARD']))
]

# Isolation des 20 Top Marques restantes
top_brands = df_filtered['brand'].value_counts().head(20).index
df_top = df_filtered[df_filtered['brand'].isin(top_brands)]


# --- 2. CRÉATION DE LA FIGURE GLOBALE (2 panneaux) ---
# On crée la fenêtre globale (légèrement plus large pour aérer le texte)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))


# --- PANNEAU GAUCHE (ax1) : LA HEATMAP ---
rfm_brand_matrix = pd.crosstab(df_top['RFM_Segment_ID'], df_top['brand'], normalize='index') * 100

# On assigne explicitement le dessin au panneau de gauche avec ax=ax1
sns.heatmap(rfm_brand_matrix, cmap='Oranges', annot=False, linewidths=.5, ax=ax1)

# On utilise les commandes spécifiques à ax1 (et non plt.title)
ax1.set_title("Brand Affinities by RFM Segment (Excluding Own Brand & Gift Cards)", fontsize=14, fontweight='bold')
ax1.set_xlabel("Top 20 Partner Brands", fontsize=12)
ax1.set_ylabel("RFM Segment (1 to 9)", fontsize=12)

# La correction de l'inclinaison appliquée UNIQUEMENT au panneau de gauche
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right', rotation_mode='anchor')


# --- PANNEAU DROIT (ax2) : LE SCATTER PLOT ---
brand_matrix = df_top.groupby('brand').agg(
    avg_rfm=('RFM_Segment_ID', 'mean'),
    avg_spend=('salesVatEUR', 'mean'),
    total_customers=('anonymized_card_code', 'nunique')
).reset_index()

# On assigne explicitement le dessin au panneau de droite avec ax=ax2
sns.scatterplot(
    data=brand_matrix, 
    x='avg_rfm', 
    y='avg_spend', 
    size='total_customers', 
    sizes=(100, 1000), 
    alpha=0.7, 
    color='darkorange',
    ax=ax2
)

# Ajout des noms des marques
for i in range(brand_matrix.shape[0]):
    ax2.text(brand_matrix['avg_rfm'].iloc[i], 
             brand_matrix['avg_spend'].iloc[i] + 1.2, # Décalage vers le haut
             brand_matrix['brand'].iloc[i], 
             horizontalalignment='center', size=8, color='black')

# Esthétique spécifique à ax2
ax2.set_title("Strategic Brand Positioning", fontsize=14, fontweight='bold')
ax2.set_xlabel("Average RFM Segment (Loyalty)", fontsize=12)
ax2.set_ylabel("Average Spend (EUR)", fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.5)
ax2.get_legend().remove() # Cache la légende pour aérer


# --- 3. FINALISATION ET SAUVEGARDE GLOBALE ---
# On gère l'espacement et on sauvegarde l'image entière une seule fois à la toute fin
plt.tight_layout()
plt.savefig('combined_dashboard_brands.png', dpi=300, bbox_inches='tight')
plt.show()
```
